1. –í—ã—á–∏—Å–ª–∏—Ç—å recovery –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ train –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery (—ç—Ç–∞–ø —Ñ–ª–æ—Ç–∞—Ü–∏–∏).
2. –ù–∞–π—Ç–∏ MAE –º–µ–∂–¥—É —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞.


–≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ `(C * (F-T)) / (F * (C - T)) * 100%`, –≥–¥–µ:

* C ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
* F ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
* T ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏.

## –ú–æ–¥–µ–ª—å

###  –ù–∞–ø–∏c–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE

```python
# –º–µ—Ç—Ä–∏–∫–∞ –¥–ª—è –æ–¥–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã
def smape(true, predict):
    return 1 / len(true) * sum( abs(true - predict) / ( (abs(true) + abs(predict)) * 0.5) ) * 100
```

```python
# –∏—Ç–æ–≥–æ–≤–∞—è –º–µ—Ç—Ä–∏–∫–∞
def final_smape(smape_r, smape_f):
    return 0.25 * smape_r + 0.75 * smape_f

```
