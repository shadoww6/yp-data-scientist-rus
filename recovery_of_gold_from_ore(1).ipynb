### Проверка расчётов эффективности обогащения

**Задачи**:
1. Вычислить recovery на обучающей выборке train для признака rougher.output.recovery (этап флотации).
2. Найти MAE между расчётами и значением признака.


Эффективность обогащения рассчитывается по формуле `(C * (F-T)) / (F * (C - T)) * 100%`, где:

* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.


```python
#так как пропуски мы ещё не обработали, исключим их из расчётов

C = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.concentrate_au']
F = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.input.feed_au']
T = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.tail_au']

rougher_recovery = C * (F - T) / (F * (C - T)) * 100
rougher_recovery_train = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.recovery']
```

```python
MAE = mean_absolute_error(rougher_recovery_train, rougher_recovery)
print(MAE)
```

Значение средней абсолютной ошибки стремится к нулю, это значит, что показатель эффективности обогащения после этапа флотации рассчитан верно.

###  Напиcание функции для вычисления итоговой sMAPE

```python
# метрика для одной величины
def smape(true, predict):
    return 1 / len(true) * sum( abs(true - predict) / ( (abs(true) + abs(predict)) * 0.5) ) * 100
```

```python
# итоговая метрика
def final_smape(smape_r, smape_f):
    return 0.25 * smape_r + 0.75 * smape_f

```
